```python
import matplotlib.pyplot as plt
import numpy as np

Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

plt.plot(Temp,Res)

plt.grid()

plt.show
plt.title("Thermistor Characteristices")

plt.xlabel("Temperature")
plt.ylabel("Resistance")



```

## BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Karnataka","GOA","TN","Maha","AP"])
y = np.array([85,26,78,30,15])

plt.bar(x,y)
plt.show()

plt.title("Literacy Rate of Diff States of INDIA")
plt.xlabel("States")
plt.ylabel("Literacy Rate")

plt.pie(y)
plt.show()
```
